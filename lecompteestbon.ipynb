# Le compte est bon
Le compte est bon est un jeu qui introduit une collection de nombres et un résultat attendu.
Ce jeu consiste à chercher une séquence de calculs qui permet d'obtenir un résultat qui est égal ou le plus proche possible d'un résultat attendu.
Cette séquence de calculs doit être élaborée avec les nombres de la collection de nombres et avec des opérations parmi +, -, x et /.
